11) Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
import pandas as pd

columns = 'age,workclass,fnlwgt,education,education-num,marital-status,occupation,relationship,race,sex,capital-gain,capital-loss,hours-per-week,native-country,income'.split(',')

df = pd.read_csv('data/adult.csv.gz', na_values='?', names=columns)

men_df = df[df['sex'] == 'Male']

married_men = men_df[men_df['marital-status'].str.startswith('Married')]
single_men = men_df[~men_df['marital-status'].str.startswith('Married')]

married_rich_ratio = (married_men[married_men['income'] == '>50K'].shape[0]) / married_men.shape[0]
single_rich_ratio = (single_men[single_men['income'] == '>50K'].shape[0]) / single_men.shape[0]

if married_rich_ratio > single_rich_ratio:
    print(f'Женатые мужчины зарабатывают больше: {married_rich_ratio:.2%} против {single_rich_ratio:.2%} у холостых.')
else:
    print(f'Холостые мужчины зарабатывают больше: {single_rich_ratio:.2%} против {married_rich_ratio:.2%} у женатых.')
```

13) Постройте сводную таблицу для сравнения уровня образования и разницы между capital.gain и capital.loss (то есть пассивный доход). Сделайте так, чтобы capital.gain - capital.loss создали не больше 10 разных уникальных значений. То есть, gain и loss - это числа в каком-то диапазоне. Это количественные признаки из вещественных чисел. С помощью математических функций, типа log, извлечение корня N-ой степени и округления можно перейти к категориальным признакам.

```python
import pandas as pd
import numpy as np

columns = 'age,workclass,fnlwgt,education,education-num,marital-status,occupation,relationship,race,sex,capital-gain,capital-loss,hours-per-week,native-country,income'.split(',')

df = pd.read_csv('data/adult.csv.gz', na_values='?', names=columns)

df['capital-gain'] = pd.to_numeric(df['capital-gain'], errors='coerce')
df['capital-loss'] = pd.to_numeric(df['capital-loss'], errors='coerce')

df['capital_diff'] = df['capital-gain'] - df['capital-loss']

df['capital_diff_log'] = np.log1p(df['capital_diff'].clip(lower=0))

df['capital_category'] = pd.cut(df['capital_diff_log'], bins=10, labels=False)

pivot_table = pd.pivot_table(df, values='capital_diff', index='education', columns='capital_category', aggfunc='count', fill_value=0)

pivot_table
```

14) Женщины из каких стран получают в среднем большую зарплату (>50K) чаще.

```python
import pandas as pd

columns = 'age,workclass,fnlwgt,education,education-num,marital-status,occupation,relationship,race,sex,capital-gain,capital-loss,hours-per-week,native-country,income'.split(',')

df = pd.read_csv('data/adult.csv.gz', na_values='?', names=columns)

women_df = df[df['sex'] == 'Female']

high_income_women = women_df[women_df['income'] == '>50K'].groupby('native-country').size()
total_women = women_df.groupby('native-country').size()

income_ratio = (high_income_women / total_women).sort_values(ascending=False)

income_ratio
```
